## Attributes
You can create it with the dot notition in the code.
BUT
easier if we abstract it, inside of class.

For this we need a **constructor**. This specifies what should happen when the class in initialized as an object as starting
To do this, we need: a function called **__init__**.
It is used to initialize the starting attributes. This is called each time we create an object.

*self* it is refering to the Blueprint itself, and the data stored within it.
Howver if we want to add attributes inside, we need to pass them as paramteres in the init funciton. and make the *self.attribute = attribute*

```python
class User:
    #pass
    def __init__(self, user_id, username):
        self.id = user_id # Not needed by defail the objec tto work, but it is easier to abbstrcut
        self.username = username # Same as above
        self.followers = 0 # Can also add default value without passing in as arguments
        self.following = 0

    def follow(self, user):
        user.followers += 1 # This refers to the other user, that this object follows, hence the user
        self.following += 1 # This refers to the user following the other one, hence the self

user_1 = User("001", "angela") # By defailt the arguments can be empty, the boject will be created, but since we abstructed the user_id, we need to pass it as parameter
#user_1.id = "001" - This is also a valid attribute assignment, but since now the init call also contains it, we don't need it
# user_1.username = "angela" Same as above
user_2 = User("002", "jack")
print(user_1.id)
```

## Methods
This is what does the boject does. 
We also also change the attributes wiht these.
Thje method is always need to have the *self* parameter in the begining

```python
user_1.follow(user_2)
print(user_1.followers)
print(user_1.following)
print(user_2.followers)
print(user_2.following)
```
